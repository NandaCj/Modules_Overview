# Problem Summary 

- This is a binary classication problem , classify if the patient is tumor is either malignant or benign 

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
breast_cancer_dataset = 'breast_cancer_dataset.csv'
```

# Dataset - Overview

- This dataset has various factors noted in the image of the breast cell

```python
df = pd.read_csv(breast_cancer_dataset)
df.head()
```

# First look shows that the data is mostly numerical and has different scales

```python
df.describe()
```

# This is a binary classification Problem , only 2 classes are listed in categorical form 

- Replacing the categorical binary output class into numerical class

Malignant = 1 
Benign = 0

```python
df['diagnosis'].value_counts()
```

```python
df = df.replace({'M':1, 'B':0})
df['diagnosis'].value_counts()
```

# Removing irrelavant Columns and Splitting X and Y 
- Y is the column diagnosis values (malignant = 1 and benign = 0)

```python
y = df.diagnosis
unwanted_columns = ['id', 'Unnamed: 32', 'diagnosis']
X = df.drop(unwanted_columns, axis=1)
```

# Applying Model without Feature Engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

# Splitting Train and Test sets

```python
x_train, x_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=42)
```

# 1. Random Forest , n_estimators 

```python
RF_Clf = RandomForestClassifier(n_estimators=100, random_state=43)
rf_clf = RF_Clf.fit(x_train, y_train)
predicted_results = rf_clf.predict(x_test)
ac = accuracy_score(y_test, predicted_results)
cm = confusion_matrix(y_test, predicted_results)
print("Accuracy :{}".format(ac))
print("Confusion Matrix Result : \n {}".format(cm))
```
